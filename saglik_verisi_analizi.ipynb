```python
import pandas as pd
import joblib
```

```python
df_test=pd.read_csv("Testing.csv")
df_train=pd.read_csv("Training.csv").drop(columns=["Unnamed: 133"])
```

```python
df_test.head()
```

```python
X_train = df_train.drop('prognosis', axis=1)
X_test = df_test.drop('prognosis', axis=1)

y_train = df_train['prognosis']
y_test = df_test['prognosis']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree_model= DecisionTreeClassifier()
```

```python
dtree_model.fit(X_train,y_train)
```

```python
predictions= dtree_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(classification_report(y_test, predictions))

accuracy = accuracy_score(y_test, predictions)
print(f"Model Doğruluk Oranı: %{accuracy * 100:.2f}")
```

```python
joblib.dump(dtree_model, 'hastalik_tahmin_modeli.pkl')

joblib.dump(X_train.columns.tolist(), 'semptom_listesi.pkl')

print("Model ve semptom listesi başarıyla .pkl olarak kaydedildi!")
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(X_train, min_support=0.05, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

top_rules = rules.sort_values(by='confidence', ascending=False)

top_rules.to_csv('birliktelik_kurallari.csv', index=False)

print("Apriori ile Birliktelik Kuralları başarıyla hesaplandı ve 'birliktelik_kurallari.csv' olarak kaydedildi!")

top_rules.head(10)
```
